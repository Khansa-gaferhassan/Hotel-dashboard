```python
x = np.linspace(0,10,100)
y = np.sin(x)
```

```python
plt.plot(x,y,label ="Sine Wave",color='blue')
plt.title("Example ")
plt.grid()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

```python

fig ,axs =plt.subplots(2,1)
axs[0].plot(x,y,'r')
axs[0].set_title('sinjnhkbjklk;ave')
axs[1].plot(x,np.cos(x),'g')
axs[1].set_title('cosine  wave')
plt.tight_layout()
plt.show()
```
